```python
# Paso 1: Importamos librerías para análisis de datos y visualización
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Paso 2: Se lee el archivo CSV y se convierte en un DataFrame
df = pd.read_csv('financials.csv', sep = ',')
df.head(5)
```

```python
sns.scatterplot(data=df, x='Price', y='Dividend Yield', hue='Sector')
plt.show()
```

```python
sns.barplot(data=df, x='Dividend Yield', y='Sector', hue='Sector', legend=False)
plt.show()
```
