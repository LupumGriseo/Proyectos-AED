Para el ejercicio "Taller 3. Comparte tu Análisis bivariado", escogí los campos 'EBITDA', 'Dividend Yield' y 'Sector'. El primero corresponde a las ganancias de una empresa antes de descontar el pago de intereses, impuestos, depreciación y amortización. El EBITDA se usa como medida de rentabilidad de una empresa a nivel operativo, ya que no incluye el efecto de los costos por deudas, impuestos y mantenimiento de activos. El segundo se refiere a la razón dividendo/precio, la cual es el dividendo pagado por una acción dividido por el precio de una acción. Por último, se consideró el campo 'Sector' para analizar la relación bivariada en los diferentes sectores de la economía de los que hacen parte las empresas analizadas.

Análisis bivariado: 
De la matriz de correlaciones calculada utilizando el método por defecto de Pearson, se puede ver que la relación entre el EBITDA y el Dividend Yield, aunque es positiva, es débil (0.126133). Esto indica que aunque las ganancias operativas de una empresa sean mayores esto no se va a traducir en un aumento de la rentabilidad de los dividendos significativo (dividendo por acción/precio por acción). Haciendo un análisis univariado sobre los campos EBITDA y Dividend Yield y diferenciando entre los sectores económicos de los que hacen parte las empresas, podemos ver que el sector de Servicios de Telecomunicaciones tiene un EBITDA y rentabilidad del dividendo promedio superior a otros sectores, por lo que sería aconcejable explorar más a fondo este sector ya que es el sector más atractivo para invertir a primera vista, puesto que el precio que se pague por acción va a 'rentar' más, ya que en promedio se van a recibir mayores dividendos por acción comprada.
